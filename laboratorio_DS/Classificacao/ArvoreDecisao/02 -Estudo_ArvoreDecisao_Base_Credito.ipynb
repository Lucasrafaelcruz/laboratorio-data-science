## Estudo 02 - Classificação Credit Data

##### Importacao Bliblioteca 

```python
import pandas as pd

from sklearn.preprocessing import LabelEncoder

import pickle as pk

from sklearn.tree import DecisionTreeClassifier
```

##### Carregamento dos Dados 

```python
with open(r'content\02.credito\credit.pkl', 'rb') as f:
    X_credit_treinamento, y_credit_treinamento, X_credit_teste, y_credit_teste = pk.load(f)
```

```python
#Ver o tamanho da amostra
X_credit_treinamento.shape,y_credit_treinamento.shape
```

```python
X_credit_treinamento
```

```python
y_credit_treinamento
```

```python
#instacia o modelo no criterio de calculo gini na variavel 
#treina a variavel com os dados do arquivo que foi colocado no pickle
arvore_credit = DecisionTreeClassifier(criterion='gini',random_state=0)
arvore_credit.fit(X_credit_treinamento, y_credit_treinamento)
```

```python
#Faz as previsoes na amostra de teste
previsoes = arvore_credit.predict(X_credit_teste)
previsoes
```

```python
# 1 nao paga
# 0 paga 
```

```python
y_credit_teste
```

**Metricas**

```python
from sklearn.metrics import accuracy_score, classification_report,confusion_matrix
```

```python
accuracy_score(y_credit_teste,previsoes)
```

```python
confusion_matrix(y_credit_teste,previsoes)
```

Isso quer dizer que 430 clientes que pagam foram classificados corretamente
6 clientes que pagaram foram classificados como nao pagam
61 clientes que nao pagam, foram classificados corretamente como nao pagam
3 clientes que nao pagam foram classificados como que pagam

```python
print(classification_report(y_credit_teste,previsoes))
```

Linha 0 - **Recall**: Ele identifica 99% dos clientes que pagam, quando ele classifica ele acerta 99% das vezes (**Precision**)

Linha 1 - **Recall**: Ele identifica 94% dos clientes que nao pagam, quando ele classifica ele acerta 91% das vezes (**Precision**)

```python
from sklearn import tree
import matplotlib.pyplot as plt

# Nome das variáveis utilizadas para treinar a árvore
previsores = ['income', 'age', 'loan']

# Cria a área (figura) onde a árvore será desenhada
fig, axes = plt.subplots(
    nrows=1,          # quantidade de linhas de gráficos
    ncols=1,          # quantidade de colunas de gráficos
    figsize=(20,20)  # tamanho da figura (largura, altura)
)

# Desenha a árvore de decisão
tree.plot_tree(
    arvore_credit,                              # modelo treinado
    feature_names=previsores,                  # nomes das variáveis
    class_names=[str(i) for i in arvore_credit.classes_],  # nomes das classes (0 e 1)
    filled=True                                # colore os nós conforme a classe predominante
)

# Salva a árvore como uma imagem PNG
fig.savefig('arvore_credit.png')

# Exibe a árvore na tela
plt.show()
```
