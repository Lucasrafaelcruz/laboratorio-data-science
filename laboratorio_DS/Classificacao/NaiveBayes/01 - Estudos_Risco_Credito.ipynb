**Base de Risco Credito**

##### Importação das Blibliotecas

```python
import pandas as pd

from sklearn.preprocessing import LabelEncoder

import pickle
```

```python
#Leitura da Base
base_risco_credito = pd.read_csv(r'content\risco_credito\risco_credito.csv')
base_risco_credito.head()
```

```python
# Criando variaveis preditoras e previsoras
X_risco_credito = base_risco_credito.iloc[:,0:4]
Y_risco_credito = base_risco_credito.iloc[:,4]
```

```python
#Features
X_risco_credito.head()
```

```python
#Variavel Resposta
Y_risco_credito.head()
```

```python
# Importar o módulo da biblioteca LabelEncoder


# Salvar os objetos LabelEncoder de cada coluna
encoders = {}

# Salvar os mapeamentos categoria -> número
mapeamentos = {}

# Loop para percorrer todas as features e transformar seus domínios em labels numéricos
for coluna in X_risco_credito.columns:

    # Instancia um LabelEncoder para a coluna atual
    encoder = LabelEncoder()

    # Aprende as categorias da coluna e realiza a transformação para números
    X_risco_credito[coluna] = encoder.fit_transform(
        X_risco_credito[coluna]
    )

    # Salva o encoder para reutilização futura
    encoders[coluna] = encoder

    # Cria um dicionário para documentar o mapeamento realizado
    mapeamentos[coluna] = {}

    # Percorre as categorias aprendidas e registra o número atribuído a cada uma
    for indice, valor in enumerate(encoder.classes_):
        mapeamentos[coluna][valor] = indice

# Exibe todos os mapeamentos realizados
print(mapeamentos)
```

```python
# Salva os dados pré-processados (X e Y) em um arquivo .pkl
# para reutilização futura sem precisar executar novamente
# toda a etapa de tratamento e codificação dos dados.

with open(r'content\risco_credito\risco_credito.pkl', 'wb') as f:
    pickle.dump([X_risco_credito, Y_risco_credito], f)
```

**Naive Bayes**

```python
from sklearn.naive_bayes import GaussianNB
```

```python
naive_risco_credito = GaussianNB()
naive_risco_credito.fit(X_risco_credito,Y_risco_credito)
```

```python
print(mapeamentos)
```

```python
# história de crédito
# boa = 0
# desconhecida = 1
# ruim = 2

# dívida
# alta = 0
# baixa = 1

# garantias
# adequada = 0
# nenhuma = 1

# renda
# 0_15 = 0
# 15_35 = 1
# acima_35 = 2

previsao = naive_risco_credito.predict([
    
    # Cliente 1
    # História: boa
    # Dívida: baixa
    # Garantias: adequada
    # Renda: acima_35
    [0, 1, 0, 2],

    # Cliente 2
    # História: ruim
    # Dívida: alta
    # Garantias: nenhuma
    # Renda: 0_15
    [2, 0, 1, 0],

    # Cliente 3
    # História: desconhecida
    # Dívida: baixa
    # Garantias: nenhuma
    # Renda: 15_35
    [1, 1, 1, 1],

    # Cliente 4
    # História: boa
    # Dívida: alta
    # Garantias: adequada
    # Renda: 15_35
    [0, 0, 0, 1],

    # Cliente 5
    # História: ruim
    # Dívida: baixa
    # Garantias: adequada
    # Renda: acima_35
    [2, 1, 0, 2]
])

print(previsao)
```

```python
previsao
```

```python
# classes_
# Mostra quais são as classes que o modelo aprendeu.
# No problema de risco de crédito geralmente:
# alto = risco alto
# moderado = risco moderado
# baixo = risco baixo
#
# Exemplo:
# ['alto' 'baixo']
print(naive_risco_credito.classes_)
```

```python
# class_count_
# Quantidade de registros de cada classe utilizados no treinamento.
#
# Exemplo:
# [7. 7.]
#
# Significa que:
# 7 exemplos pertencem à primeira classe
# 7 exemplos pertencem à segunda classe
print(naive_risco_credito.class_count_)
```

```python
# class_prior_
# Probabilidade inicial (a priori) de cada classe.
#
# É calculada antes de analisar as características
# do cliente (história, dívida, garantias e renda).
#
# Exemplo:
# [0.5 0.5]
#
# Significa:
# 50% dos registros são da primeira classe
# 50% dos registros são da segunda classe
print(naive_risco_credito.class_prior_)
```
