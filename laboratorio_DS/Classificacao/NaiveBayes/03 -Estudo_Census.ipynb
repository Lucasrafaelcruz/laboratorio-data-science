## 📊 Análise do Modelo Gaussian Naive Bayes - Census Income

### 📌 Distribuição dos dados de teste (y_teste)

* 75,6% dos indivíduos possuem renda menor ou igual a 50 mil dólares (3.693 de 4.885)
* 24,4% dos indivíduos possuem renda superior a 50 mil dólares (1.192 de 4.885)

Isso demonstra um cenário desbalanceado, onde a maior parte da população pertence à faixa de renda até 50 mil dólares.

---

## 🧠 Matriz de Confusão

| Real \ Previsto | <=50K | >50K |
| --------------- | ----- | ---- |
| <=50K           | 1172  | 2521 |
| >50K            | 35    | 1157 |

---

## ✅ Classe <=50K — Renda até 50 mil dólares

O modelo apresentou um comportamento bastante conservador ao realizar previsões dessa classe.

* **Precision: 97%**

Quando o modelo prevê que uma pessoa possui renda de até 50 mil dólares, ele está correto na grande maioria das vezes.

* **Recall: 32%**

O modelo conseguiu identificar apenas 32% das pessoas que realmente pertencem a essa faixa de renda.

📌 **Interpretação:**

Embora as previsões sejam altamente confiáveis, o modelo deixa de reconhecer grande parte dos indivíduos dessa classe.

➡️ Aproximadamente 68% das pessoas que realmente possuem renda até 50 mil dólares foram classificadas incorretamente como pertencentes à faixa de renda superior.

---

## ⚠️ Classe >50K — Renda superior a 50 mil dólares

Esta foi a classe favorecida pelo modelo durante as previsões.

* **Precision: 31%**

Quando o modelo prevê que uma pessoa possui renda superior a 50 mil dólares, ele acerta apenas 31% das vezes.

* **Recall: 97%**

Por outro lado, o modelo consegue encontrar praticamente todas as pessoas que realmente pertencem a essa faixa de renda.

📌 **Interpretação:**

O modelo identifica quase todos os indivíduos de alta renda, porém gera uma grande quantidade de falsos positivos.

➡️ Muitas pessoas com renda inferior a 50 mil dólares são classificadas incorretamente como pertencentes à faixa de renda superior.

---

## 📈 Visão Geral do Modelo

* **F1-Score (<=50K): 48%**
* **F1-Score (>50K): 48%**
* **Accuracy Geral: 48%**

O F1-Score evidencia que o modelo não conseguiu equilibrar adequadamente precisão e recall para nenhuma das classes.

Embora o recall da classe >50K seja muito elevado, a baixa precisão reduz significativamente a qualidade das previsões.

Da mesma forma, a classe <=50K apresenta excelente precisão, mas um recall extremamente baixo.

---

## 🧾 Conclusão

O Gaussian Naive Bayes apresentou forte desequilíbrio entre as classes analisadas.

O modelo demonstrou excelente capacidade de encontrar indivíduos com renda superior a 50 mil dólares, porém ao custo de classificar incorretamente um grande número de pessoas da faixa de renda inferior.

📌 Em termos práticos:

* O modelo dificilmente deixa de identificar indivíduos de alta renda 👍
* Porém classifica muitas pessoas de renda mais baixa como pertencentes à faixa de alta renda ⚠️
